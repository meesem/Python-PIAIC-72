```python
# Smart Grocery Store Assistant

product_catalog = {
    "apple": (100, 10),
    "banana": (60, 20)
}

def view_products():
    print("\n--- Available Products ---")
    for product, (price, stock) in product_catalog.items():
        print(f"{product.capitalize()} – Rs.{price}, In stock: {stock}")
    print()

def add_or_update_product():
    product = input("Enter product name: ").lower()
    price = int(input("Enter product price: Rs."))
    stock = int(input("Enter stock quantity: "))
    product_catalog[product] = (price, stock)
    print(f"{product.capitalize()} added/updated successfully!\n")

def purchase_items():
    cart = {}
    while True:
        product = input("Enter product to buy (or 'done' to finish): ").lower()
        if product == "done":
            break
        if product not in product_catalog:
            print("Product not found.")
            continue
        quantity = int(input(f"Enter quantity of {product}: "))
        price, stock = product_catalog[product]
        if quantity > stock:
            print("Not enough stock.")
            continue
        cart[product] = cart.get(product, 0) + quantity
        product_catalog[product] = (price, stock - quantity)

    if cart:
        print("\n--- Bill ---")
        total = 0
        for product, quantity in cart.items():
            price = product_catalog[product][0]
            cost = price * quantity
            total += cost
            print(f"{product.capitalize()} x{quantity} – Rs.{cost}")
        print(f"Total Bill: Rs.{total}\n")
        restock_alert()

def restock_alert():
    print("\n--- Restock Alerts ---")
    for product, (price, stock) in product_catalog.items():
        if stock < 5:
            print(f"{product.capitalize()} is low on stock! Only {stock} left.")
    print()

def main():
    print("Welcome to the Smart Grocery Store Assistant")
    while True:
        print("1. View Products")
        print("2. Add/Update Product")
        print("3. Purchase Items")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            view_products()
        elif choice == "2":
            add_or_update_product()
        elif choice == "3":
            purchase_items()
        elif choice == "4":
            print("Thank you for using the assistant. Goodbye!")
            break
        else:
            print("Invalid choice. Try again.\n")

if __name__ == "__main__":
    main()
```

project 2..Library book management system

```python
# Initial book catalog
books = {
    "1984": {"author": "Orwell", "available": True},
    "Dracula": {"author": "Stoker", "available": True}
}

# List to track borrowed books
borrowed_books = []

# Function to view all books
def view_books():
    print("\n--- Library Books ---")
    for title, info in books.items():
        status = "Available" if info["available"] else "Borrowed"
        print(f'"{title}" by {info["author"]} – {status}')
    print()

# Function to borrow a book
def borrow_book():
    book_title = input("Enter book name to borrow: ").strip().lower()
    found = False
    for title in books:
        if title.lower() == book_title:
            found = True
            if books[title]["available"]:
                books[title]["available"] = False
                borrowed_books.append(title)
                print(f'You have borrowed "{title}".\n')
            else:
                print(f'"{title}" is already borrowed.\n')
            break
    if not found:
        print("Book not found.\n")

# Function to return a book
def return_book():
    book_title = input("Enter book name to return: ").strip().lower()
    found = False
    for title in books:
        if title.lower() == book_title:
            found = True
            if not books[title]["available"]:
                books[title]["available"] = True
                borrowed_books.remove(title)
                print(f'You have returned "{title}".\n')
            else:
                print(f'"{title}" was not borrowed.\n')
            break
    if not found:
        print("Book not found.\n")

# Function to add a new book
def add_book():
    title = input("Enter new book title: ").strip()
    author = input("Enter author's name: ").strip()
    books[title] = {"author": author, "available": True}
    print(f'Book "{title}" by {author} added successfully!\n')

# Function to view borrowed books
def view_borrowed_books():
    if not borrowed_books:
        print("\nNo books currently borrowed.\n")
    else:
        print("\n--- Borrowed Books ---")
        for title in borrowed_books:
            print(f'"{title}" by {books[title]["author"]}')
        print()

# Main menu loop
def main():
    print("📚 Welcome to the Library System 📚")
    while True:
        print("1. View Books")
        print("2. Borrow Book")
        print("3. Return Book")
        print("4. Add Book")
        print("5. View Borrowed Books")
        print("6. Exit")
        choice = input("Enter your choice: ").strip()

        if choice == "1":
            view_books()
        elif choice == "2":
            borrow_book()
        elif choice == "3":
            return_book()
        elif choice == "4":
            add_book()
        elif choice == "5":
            view_borrowed_books()
        elif choice == "6":
            print("Thank you for using the Library System. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.\n")

# Run the program
if __name__ == "__main__":
    main()
```

project 3 ..student report card generator

```python
# Dictionary to store all student data
students = {}

# Function to calculate grade based on average
def calculate_grade(average):
    if average >= 90:
        return 'A'
    elif average >= 80:
        return 'B'
    elif average >= 70:
        return 'C'
    elif average >= 60:
        return 'D'
    else:
        return 'F'

# Function to add a student's report
def add_student():
    name = input("Enter student name: ").strip()
    roll = input("Enter roll number: ").strip()

    # Input marks for 5 subjects
    marks = []
    print("Enter marks for 5 subjects:")
    for i in range(1, 6):
        mark = int(input(f"Subject {i} marks: "))
        marks.append(mark)

    # Store student data in the dictionary
    students[name] = {
        "roll": roll,
        "marks": marks
    }
    print(f"{name}'s data added successfully!\n")

# Function to generate and display report cards
def generate_reports():
    if not students:
        print("No student data available.\n")
        return

    for name, data in students.items():
        roll = data["roll"]
        marks = data["marks"]
        total = sum(marks)
        average = total / len(marks)
        grade = calculate_grade(average)

        print("\n--- Report Card ---")
        print(f"Student Name: {name}")
        print(f"Roll No: {roll}")
        print(f"Marks: {marks}")
        print(f"Total: {total}")
        print(f"Average: {average:.2f}")
        print(f"Grade: {grade}")
        print("--------------------")

# Main loop
def main():
    print("🎓 Student Report Card Generator 🎓")
    while True:
        print("\n1. Add Student")
        print("2. Generate Report Cards")
        print("3. Exit")
        choice = input("Enter your choice: ").strip()

        if choice == "1":
            add_student()
        elif choice == "2":
            generate_reports()
        elif choice == "3":
            print("Thank you for using the Report Card Generator. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.\n")

# Run the program
if __name__ == "__main__":
    main()
```

bonus project-mini quiz game

```python
# List of questions with their correct answers
quiz = [
    {"question": "What is the capital of France?", "answer": "paris"},
    {"question": "What is 5 + 7?", "answer": "12"},
    {"question": "Who wrote 'Romeo and Juliet'?", "answer": "shakespeare"},
    {"question": "What is the largest planet in our solar system?", "answer": "jupiter"},
    {"question": "What color do you get by mixing red and white?", "answer": "pink"}
]

# Variable to store the user's score
score = 0

# Welcome message
print("🎉 Welcome to the Mini Quiz Game! 🎉")
print("Answer the following 5 questions:\n")

# Loop through each question in the quiz list
for i, item in enumerate(quiz, start=1):
    # Show the question
    print(f"Q{i}: {item['question']}")

    # Get user's answer and convert it to lowercase for case-insensitive matching
    user_answer = input("Your Answer: ").strip().lower()

    # Check if the user's answer matches the correct answer
    if user_answer == item["answer"]:
        print("✅ Correct!\n")
        score += 1  # Increase score if correct
    else:
        print(f"❌ Wrong! The correct answer was: {item['answer'].capitalize()}\n")

# Show final score and message
print(f"🏁 Final Score: {score}/{len(quiz)}")

# Feedback based on score
if score == len(quiz):
    print("🌟 Excellent job!")
elif score >= 3:
    print("👍 Good effort!")
else:
    print("📘 Keep practicing!")
```
